```python
stopwords=stopwords.words('english')
```

```python
def clean_text(line):

    #converting it to lower
    line=line.lower()
    print(line)

    tokens=word_tokenize(line)
    print(tokens)

    #removing the punctuations
    words=[w for w in tokens if w not in string.punctuation]
    print(words)

    #removing the stopwords
    words=[w for w in words if w not in stopwords]
    print(words)
    

    return ' '.join(words)
```

```python
clean_text("this product is very nice and useful!!!")
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:clean_text(x))
```
