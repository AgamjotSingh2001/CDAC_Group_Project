```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from nltk.stem.wordnet import WordNetLemmatizer
import os
path = os.getcwd()
```

```python
df = pd.read_json(r'E:\Project\Project_dataset\Toys_and_Games.json',nrows=100000,lines=True)
```

```python
df.head()
```

```python
df = df[['reviewText','class']]
df.head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
ls = WordNetLemmatizer()
stopwords = [stopwords.words('english')]
```

```python
def clean_text(sent):
    token1 = word_tokenize(sent) #tokenizing the sentences
    #token2 = [x.lower() for x in token1 if x not in string.punctuation] #Removing the punctuations
    token2 = [x.lower() for x in token1 if x.isalpha() or x.isdigit()] #Removing the punctuations
    token3 = [ls.lemmatize(x) for x in token2 if x not in stopwords] #removing affixes
    return token3 
```

```python
y=df['class']
```

```python
vectorizer = TfidfVectorizer(max_features=10000,analyzer=clean_text)  # You can adjust the number of features as needed
x = vectorizer.fit_transform(df['reviewText'])
x
```

```python
from sklearn.model_selection import train_test_split
train_sent,test_sent,train_labels,test_labels=train_test_split(x,y,stratify=y,test_size=0.2,random_state=0)
```

```python
model = GaussianNB()
```

```python
model.fit(train_sent.toarray(),train_labels)
```

```python
import joblib
joblib.dump(model, 'Naive_bayes.joblib')
```

```python
y_pred = model.predict(test_sent.toarray())
```

```python
y_pred
```

```python
print(accuracy_score(test_labels,y_pred))
```

```python
print(classification_report(test_labels,y_pred))
```
