```python
word2vec_model = Word2Vec(window=5,min_count=2,vector_size=300)
```

```python
max_sequence_length=300
```

```python
# Convert words to Word2Vec embeddings
embedding_matrix = np.zeros((len(word2vec_model.wv.key_to_index) + 1,word2vec_model.vector_size))
for word, i in word2vec_model.wv.key_to_index.items():
    embedding_vector = word2vec_model.wv[word]
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```

```python
embedding_matrix.shape
```

```python
# Convert text data to sequences of Word2Vec indices
def convert_to_indices(sentence_list):
    sequences = []
    for sentence in sentence_list:
        sequence = []
        for word in sentence:
            if word in word2vec_model.wv:
                sequence.append(word2vec_model.wv.key_to_index[word])
        sequences.append(sequence)
    return sequences
```

```python
train_seq = convert_to_indices(train_list)
test_seq = convert_to_indices(test_list)
```

```python
# Pad sequences to ensure uniform length
padded_train_sequences = pad_sequences(train_seq, maxlen=max_sequence_length, padding='post')
padded_test_sequences = pad_sequences(test_seq, maxlen=max_sequence_length, padding='post')
```

```python
padded_train_sequences
```

```python
padded_test_sequences
```

```python
lstm_model=tf.keras.models.Sequential()
lstm_model.add(tf.keras.layers.Embedding(
                    input_dim=len(word2vec_model.wv.key_to_index) +  1,
                    output_dim=word2vec_model.vector_size, 
                    weights=[embedding_matrix],
                    input_length=max_sequence_length,
                    trainable=False))
lstm_model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)))
lstm_model.add(tf.keras.layers.Dense(32,activation='relu'))
lstm_model.add(tf.keras.layers.Dense(16,activation='relu'))
lstm_model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
lstm_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
lstm_model.fit(padded_train_sequences,train_df['class'],epochs=5,validation_data=[padded_test_sequences,test_df['class']],batch_size=64)
```

```python
#sentence prediction
sent = """Firstly product’s dimensions is very small against as shown in the description.
Secondly product’s design is not a appropriate to be used as per intended. Objects of all shapes can be fitted in other shapes. Child cannot match the object to its shape which defeats the purpose of this product."""
```

```python
def clean_text_for_prediction(sent):
    sent_list=simple_preprocess(sent)
    sequence = []
    for word in sent_list:
        if word in word2vec_model.wv:
            sequence.append(word2vec_model.wv.key_to_index[word])
    return sequence
```

```python
sent_seq = clean_text_for_prediction(sent)
padded_sequences = pad_sequences([sent_seq], maxlen=max_sequence_length, padding='post')
```

```python
padded_sequences
```

```python
res=lstm_model.predict(padded_sequences)
if res>0.5:
    print('spam')
else:
    print('non spam')
```
