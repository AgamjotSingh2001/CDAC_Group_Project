```python
import tensorflow as tf
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('D:\CDAC_PUNE_PROJECT\\toys.csv',index_col=0)
```

```python
df.head()
```

```python
cleaned=[]
def clean_text(sent):

    cleaned.append(simple_preprocess(sent))
df['reviewText'].apply(lambda x:clean_text(x))
```

```python
cleaned
```

```python
train_df,test_df=train_test_split(df,stratify=df['class'],test_size=0.2,random_state=0)
```

```python
cleaned=[]
```

```python
train_df['reviewText'].apply(lambda x:clean_text(x))
```

```python
train_list=cleaned
```

```python
train_list
```

```python
model = Word2Vec(window=5,min_count=2,vector_size=300)
```

```python
model.build_vocab(train_list)
```

```python
model.train(train_list,total_examples=model.corpus_count,epochs=model.epochs)
```
