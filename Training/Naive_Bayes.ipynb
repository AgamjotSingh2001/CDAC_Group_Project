```python
# Import Required Libraries and Functions


import numpy as np

import pandas as pd 

import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.naive_bayes import MultinomialNB, BernoulliNB

import string
```

```python
# Stopwords to remove redundunt words. 
stopwords = stopwords.words("english")

# Punctuations To remove punctuations from the reviews.
# punctuations = string.punctuation
```

```python
# As we need each row tokenized we did Data Preprocessing.
# find these steps in Data_Preprocessing.ipynb file
# Read only 3 lakh rows from Dataset.
toys = pd.read_csv("D:\\Desktop\\toys_data.csv",nrows=300000)
```

```python
# First five Rows From Dataset

toys.head()
```

<h5>we need to convert the words to their base form
because if we don't convert these words then both the words(with affixes and without affixes) 
will have different frequency when vectorized</h5>


```python
# Initialize lemmatizer (create an object)

wnl = WordNetLemmatizer()
```

```python
def clean_text(lst):
    '''This function takes input as list of words(each row)
        and removes punctuations and stopwords. 
    '''
    # Remove Punctuations and redundent symbols.
    token1 = [word.lower() for word in lst if word.lower().isalpha() or word.lower().isdigit()]
    
    # Remove Punctuations
    # token2 = [word for word in token1 if word not in punctuations]
    
    # Convert words to their root form.
    token3 = [wnl.lemmatize(words) for words in token1]

    return token3
```

```python
# Divide the dataset into review and class

X = toys['reviewText'][0:100000]
y = toys['class'][0:100000]
```

```python
# Vectorizing the data using TfIdf Vectorizier.
from sklearn.feature_extraction.text import TfidfVectorizer

# Initiating the TfIdf Vectorizer
# Analyzer will take care of all the preprocessing.
tfidf = TfidfVectorizer(analyzer=clean_text)

# Fiting data to the vectorizer 
X_vect = tfidf.fit_transform(X)

# Checking the Dimentionality.
X_vect
```

```python
# Split the data into training and testing with seed value 24.
# 90% training data and 10% testing.  
X_train,X_test,y_train,y_test = train_test_split(X_vect,y,random_state=24,test_size=0.2)
```

```python
# Initialize the MultinomialNB model.
mnb = MultinomialNB()

# Train the multinomialNB model.
mnb.fit(X_train,y_train)
```

```python
# predict on Test Data.

y_pred_mnb = mnb.predict(X_test)
```

```python
# Confusion matrix to evaluate models classification.

print(confusion_matrix(y_pred_mnb,y_test))
```

```python
# Accuracy score to evaluate models performance.

print(accuracy_score(y_pred_mnb,y_test))
```

```python
# Initialize BernoulliNB Model.
bnb = BernoulliNB()

# Train the BernoulliNB Model.
bnb.fit(X_train,y_train)
```

```python
# Predict on test Data.
y_pred_bnb = bnb.predict(X_test)
```

```python
# Confusion matrix to evaluate models classification.

print(confusion_matrix(y_pred_bnb,y_test))
```

```python
# Accuracy score to evaluate models performance.

print(accuracy_score(y_pred_bnb,y_test))
```
